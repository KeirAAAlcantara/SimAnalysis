```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("spatial_awareness.csv")

df.head()
```

```python
df["Status"].unique()
```

```python
total = df["Status"].count()
correct = (df["Status"] == "correct").sum()
incorrect = (df["Status"] == "incorrect").sum()
no_answer = (df["Status"] == "no_answer").sum()
```

```python
print("Number of Correct awareness: ", correct)
print("Number of incorrect or not answered: ", incorrect)
print("No Data: ", no_answer)
```

```python
print(f"Number of Correct awareness: {(correct/total)*100}%")
print(f"Number of incorrect or not answered: {(incorrect/total)*100}%")
print(f"No Data: {(no_answer/total)*100}%")
```

```python
accuracy = correct/(correct+incorrect)
print(f"Accuracy: {accuracy*100}%")
```
