Num DF and Den DF will not change across all comparisons.
- Num DF is degrees of freedom of factor = 3 (k-1)
- Den DF is degrees of freedom of error = 57 (n-1)(k-1)

n = number of participants = 20 
k = number of conditions (alerts) = 4 alerts 

F-Value measures: 
Numerator = variability between tasks
Denominator = variability within subjects (error) 
- Large F = condition means are very different (significant effect) 
    - critical F ~ 2.76, so if F>= 2.8 (significant)
- Small F = differene between means are small relative to noise 

Pr > F (p-value)
- null hypothesis true or not
- p > 0.05

**There is no significant difference in NASA-TLX raw score workload across all 4 alert types.**

The results indicate no statistically significant difference in raw score F = 1.11, p = 0.35. 

The smallest F Value is effort with a 0.18, while the highest is Frustration with 2.10. This also translates to the p-value with effort having 0.91 and frustration having 0.11. 
 Even if frustration is larger than the rest, 2.10 is still smaller than the critical value being 2.79. 

 All of the dimensions in NASA-TLX are below the critical value which tells us to **Fail to reject the Null Hypothesis**. All p-values were greater than the significance level of 0.05, that tells us that there were no statistically significant differences in the workload across the four alert types for any NASA-TLX dimension. Additionally, all the F-values were bellow the critical value (2.76). These results suggest that despite the alert type, it did not affect the participant's percieve cognitive workload.

```python
def categorize_tlx(score):
    if 0 <= score <= 9:
        return "Low"
    elif 10 <= score <= 29:
        return "Medium"
    elif 30 <= score <= 49:
        return "Somewhat High"
    elif 50 <= score <= 79:
        return "High"
    elif 80 <= score <= 100:
        return "Very High"
    else:
        return None
```

```python
def condensed_tlx(score):
    if 0 <= score <= 29:
        return "Low - Medium"
    elif 30 <= score <= 49:
        return "Somewhat High"
    elif 50 <= score <= 100:
        return "High - Very High"
    else:
        return None
```

```python
metrics = ["Mental_Demand", "Physical_Demand", "Temporal_Demand", 
           "Performance", "Effort", "Frustration"]

# Create new categorized columns
for col in metrics:
    HO[col + "_Category"] = HO[col].apply(condensed_tlx)
```

```python
# List of categorized columns
metrics = ["Mental_Demand_Category", "Physical_Demand_Category", 
           "Temporal_Demand_Category", "Performance_Category", 
           "Effort_Category", "Frustration_Category"]

# Create a table of value counts for all dimensions
summary_tableHO = pd.DataFrame()

for col in metrics:
    counts = HO[col].value_counts().reindex(["Low - Medium", "Somewhat High", "High - Very High"], fill_value=0)
    summary_tableHO[col] = counts

# Transpose for easier reading (optional)
summary_tableHO = summary_tableHO.T

print(summary_tableHO)
```
