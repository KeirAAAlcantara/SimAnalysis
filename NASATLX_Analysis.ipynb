```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cleaned_NASATLX.csv")

df.head()
```

```python
# Combining 6 columns 
df_long = df.melt(
    id_vars="Task Name", 
    value_vars=["Mental Demand", "Physical Demand", "Frustration", 
                "Performance", "Effort", "Temporal Demand"],
    var_name="Metric",
    value_name="Score"
)
```

```python
plt.figure(figsize=(12, 6))

sns.barplot(data=df_long, 
            x="Task Name", 
            y="Score", 
            hue="Metric")

plt.xticks(rotation=45)
plt.title("Task Load Metrics by Task")
plt.legend(title="Metric", bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

```python
#Mental Demand

sns.barplot(data = df, x= "Mental Demand", y= "Task Name")
plt.title('Mental Demand across all drives')
plt.xlabel('')
plt.ylabel('Alert Type')
plt.show()
```

```python
#Mental Demand

sns.barplot(data = df, x= "Temporal Demand", y= "Task Name")
plt.title('Temporal Demand across all drives')
plt.xlabel('')
plt.ylabel('Alert Type')
plt.show()
```
