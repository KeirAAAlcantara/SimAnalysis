```python
import pandas as pd
from statsmodels.stats.anova import AnovaRM
from scipy.stats import ttest_rel
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cleaned_NASATLX.csv")

display(df)
```

```python
print(df.columns)
```

**Running Repeated Measures ANOVA**

Pr > F ; if P <0.05 -> there is a significant effect of cue condition 
Answers the question: Does cue condition affect each workload component? 

- P has no significant effect on cue condition

```python
anova = AnovaRM(
    data=df,
    depvar="Mental_Demand",
    subject="Participant",
    within=["Task_Name"]
).fit()

print(anova)
```

```python
anova = AnovaRM(
    data=df,
    depvar="Physical_Demand",
    subject="Participant",
    within=["Task_Name"]
).fit()

print(anova)
```

```python
anova = AnovaRM(
    data=df,
    depvar="Temporal_Demand",
    subject="Participant",
    within=["Task_Name"]
).fit()

print(anova)
```

```python
anova = AnovaRM(
    data=df,
    depvar="Performance",
    subject="Participant",
    within=["Task_Name"]
).fit()

print(anova)
```

```python
anova = AnovaRM(
    data=df,
    depvar="Effort",
    subject="Participant",
    within=["Task_Name"]
).fit()

print(anova)
```

```python
anova = AnovaRM(
    data=df,
    depvar="Frustration",
    subject="Participant",
    within=["Task_Name"]
).fit()

print(anova)
```

### Means 

```python
df[df['Task_Name'] == "AH"] \
    .groupby('Participant')['Raw_Score'] \
    .mean()
```

```python
conditions = ["NC", "AO", "HO", "AH"]

for condition in conditions:
    subset = df[df["Task_Name"] == condition]["Raw_Score"]
    mean = subset.mean()
    sd = subset.std()
    print(f"{condition}: M = {mean:.2f}, SD = {sd:.2f}")
```

The cue type did not significantly affect perceived workload.
